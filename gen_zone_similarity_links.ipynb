```python
limit = 500
h, w = (224, 224)
circle_r = 5

cv_table = 'cv_results_225'

conn_string = "host='localhost' dbname='indoor_position' user='postgres' password='tiancai' port='5432'"
conn = psycopg2.connect(conn_string)
cur = conn.cursor()

query = '''truncate table gare_st_lazare.links; select * from gare_st_lazare.zone_lookup; '''
cur.execute(query)
results1 = cur.fetchall()
conn.commit()

count_dic = {}
data_dic  = {}
points_lookup = {}

# build a dictionary with two levels to store computed scores
for item in results1:
    count_dic[item[0]] = 0
    data_dic[item[0]] = {}
    points_lookup[item[0]] = item[2]
    for item2 in results1:
        data_dic[item[0]][item2[0]] = 0

# add value to the built dictionary
for item in results1:
    query = '''select * from gare_st_lazare.cv_results_225 where 
            category = '%s'; ''' % (item[0])
    cur.execute(query)
    results2 = cur.fetchall()
    conn.commit()
    for entry in results2:
        for i in range(5):
            data_dic[item[0]][entry[i + 2]] += math.pow(entry[i + 7], 2)
    count_dic[item[0]] += 1

# insert each of the value to database
for key1, value1 in data_dic.items():
    for key2, value2 in data_dic[key1].items():
        query = '''insert into gare_st_lazare.links values('%s', '%s', 
        ST_MakeLine(ST_AsText('%s'), ST_AsText('%s')), %s); 
        ''' % (key1, key2, points_lookup[key1], points_lookup[key2], value2 / count_dic[key1])
        print(query)
        cur.execute(query)
        conn.commit()
    
        
```
