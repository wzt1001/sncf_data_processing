```python
import numpy as np
import cv2
import matplotlib
from matplotlib import pyplot as plt
```

```python
image_id = '2'
imgL = cv2.imread('../data/test_images/HET_0006_3840x2160/%s_left.jpg' % image_id, 0)
imgR = cv2.imread('../data/test_images/HET_0006_3840x2160/%s_right.jpg' % image_id, 0)
```

```python
plt.imshow(imgL)
```

```python
if not imgL.shape == imgR.shape:
    imgR = imgR[:imgL.shape[0], :]
    
imgL = imgL[:,480:1440]
imgR = imgR[:,480:1440]
# imgR = imgR[:,1440:2400]
```

```python
print(imgL.shape)
print(imgR.shape)
```

```python
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=35)
disparity = stereo.compute(imgL,imgR)
plt.imshow(disparity,'gray')
plt.show()
```

```python
matplotlib.image.imsave('../data/test_images/HET_0006_3840x2160/%s_depth.jpg' % image_id, disparity)
```
